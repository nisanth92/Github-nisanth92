```python
# -*- coding: utf-8 -*-
import csv
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import json
import vincent

#Get all the counties and store them in a dataframe
with open('us_counties.topo.json') as json_data:
    data = json.load(json_data)
county_list=[]
for i in range(0,len(data['objects']['us_counties.geo']['geometries'])):
    county_list.append(data['objects']['us_counties.geo']['geometries'][i]['properties']['FIPS'])
county_series=Series(np.unique(county_list)).sort_values()
heat_map=DataFrame(county_series,columns=["fips"])
#Get the clusters formed using K-Means
cluster_data=pd.read_csv("Cluster_counties.csv")
#Find the index of each cluster group and assign the cluster numbers
cols0=cluster_data[cluster_data['cluster']==0].index
cols1=cluster_data[cluster_data['cluster']==1].index
cols2=cluster_data[cluster_data['cluster']==2].index
cols3=cluster_data[cluster_data['cluster']==3].index
cols4=cluster_data[cluster_data['cluster']==4].index
heat_map['clus']=0
heat_map['clus'][cols1]=1
heat_map['clus'][cols2]=2
heat_map['clus'][cols3]=3
heat_map['clus'][cols4]=4
#Use vincent module to generate the heat-map
county_topo = "us_counties.topo.json"
geo_data = [{'name': 'counties',
             'url': county_topo,
             'feature': 'us_counties.geo'}]

vis = vincent.Map(data=heat_map, geo_data=geo_data, scale=1000,
                  projection='albersUsa', data_bind='clus',
                  data_key='fips', map_key={'counties': 'properties.FIPS'})
vis.scales['color'].type = 'threshold'
#vis.rebind(column='cluster',brew='Spectral')
vis.colors(brew=None, range_=['rgb(255,255,204)','rgb(255,255,204)','rgb(0,109,44)','rgb(116,196,118)','rgb(107,174,214)','rgb(165,15,21)'])
vis.scales['color'].domain = [0,1,2,3,4]
vis.legend(title='Obesity Clusters')
vis.to_json('vega.json')
vis.display()

```
